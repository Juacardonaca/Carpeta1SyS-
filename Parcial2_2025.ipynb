```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

A = 1  # Amplitud
f0 = 60  # Frecuencia
T = 1  # Duración de la señal
fs = 1000  # Frecuencia de muestreo

t = np.linspace(0, T, int(fs * T), endpoint=False) #Vector de tiempo

x_t = (A * np.cos(2 * np.pi * F0 * t)) ** 2 #Usamos la señal expresada en el ejercicio

#calcular la transformada de Fourier discreta
x_f = fft(x_t)
freq = fftfreq(len(x_t), 1/fs) #frecuencias correspondientes

#Espectro de Fourier exponencial
magnitud = np.abs(x_f)
fase = np.angle(x_f)

magnitud_db = 20 * np.log10(magnitud)  #magnitud a decibelios
plt.figure(figsize=(12, 6)) #diagrama de bode

plt.subplot(2, 1, 1)
plt.semilogx(freq, magnitud_db)
plt.title('Diagrama de Bode')
plt.ylabel('Magnitud (dB)')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.semilogx(freq, np.degrees(fase))  # Convierte la fase a grados
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (grados)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
